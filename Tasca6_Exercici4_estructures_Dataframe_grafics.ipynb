  - Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

```python
milestokmh = 60 / 0.62137

# Average Speed

delayedFlightsRelevant['AverageSpeed'] = (delayedFlightsRelevant['Distance'] / delayedFlightsRelevant['ActualElapsedTime'])*milestokmh

delayedFlightsRelevant['AverageSpeed'].head()
```

```python
# Departure & Arrival.

delayedFlightsRelevant['FlyDelay'] = np.where((delayedFlightsRelevant['ArrDelay'] > 0) & (delayedFlightsRelevant['DepDelay'] > 0), "yes", "no")

delayedFlightsRelevant['FlyDelay'].head()
```

```python
# Delayed

delayedFlightsRelevant['Delayed'] = np.where(delayedFlightsRelevant['ArrDelay'] > 0, 1, 0)


delayedFlightsRelevant['Delayed'].head()
```

```python
# New row 'Date'

delayedFlightsRelevant.rename(columns = {'DayofMonth':'Day'}, inplace = True)

delayedFlightsRelevant['Date'] = pd.to_datetime(delayedFlightsRelevant[['Year', 'Month', 'Day']])

delayedFlightsRelevant.head()
```

- Taula del vols amb més endarreriments acumulats

```python
carriers = pd.read_csv('../../Data/carriers.csv')
carriers.set_index('Code', inplace=True)
carriers.head()
```

```python
# Merge Data
```

```python
delayedFlightsRelevant = pd.merge(delayedFlightsRelevant, carriers, how='left', left_on='UniqueCarrier', right_on='Code')
delayedFlightsRelevant.rename(columns={'Description': 'Carrier'}, inplace=True)
delayedFlightsRelevant.drop('UniqueCarrier', axis=1, inplace=True)
delayedFlightsRelevant.head()
```

```python
delays = delayedFlightsRelevant[['Delayed', 'Carrier']].copy()
delayed_carriers = delays.groupby('Carrier').aggregate(sum)
delayed_carriers.sort_values(by='Delayed', axis=0, ascending=False)
```

- Quins són els vols més llargs? I els més endarrerits?
- Etc.

```python
airports = pd.read_csv('../../Data/airports.csv')
airports.set_index('iata', inplace=True)
airports.head()
```

```python
airport_codes = airports[['airport']]
delayedFlightsRaw = delayedFlightsRaw.merge(airport_codes, how='left', left_on='Origin', right_on='iata')
delayedFlightsRaw = delayedFlightsRaw.merge(airport_codes, how='left', left_on='Dest', right_on='iata')
delayedFlightsRaw.rename(columns={'airport_x': 'OriginAirp', 'airport_y': 'DestAirp'}, inplace=True)
delayedFlightsRaw['Flight'] = delayedFlightsRaw['OriginAirp'] + ' - ' + delayedFlightsRaw['DestAirp']
delayedFlightsRaw.head()
```

- Longer Flights

```python
flights = delayedFlightsRaw[['ActualElapsedTime', 'ArrDelay', 'Flight', 'OriginAirp', 'DestAirp']].copy()
longer_flights = flights.groupby('Flight')
longer_flights.mean().sort_values(by='ActualElapsedTime', axis=0, ascending=False).head()
```

- More delayed

```python
more_delayed = flights.groupby('Flight')
more_delayed.mean().sort_values(by='ArrDelay', axis=0, ascending=False).head()
```
