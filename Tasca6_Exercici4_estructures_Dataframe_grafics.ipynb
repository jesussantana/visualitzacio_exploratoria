  - Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

```python
milestokmh = 60 / 0.62137

# Departure & Arrival.

delayedFlightsRelevant['DoubleDelay'] = np.where((delayedFlightsRelevant['ArrDelay'] > 0) & (delayedFlightsRelevant['DepTime'] > 0), "yes", "no")

# Delayed

delayedFlightsRelevant['Delayed'] = np.where(delayedFlightsRelevant['ArrDelay'] > 0, "yes", "no")


# New row 'Date'

delayedFlightsRelevant.rename(columns = {'DayofMonth':'Day'}, inplace = True)

delayedFlightsRelevant['Date'] = pd.to_datetime(delayedFlightsRelevant[['Year', 'Month', 'Day']])

delayedFlightsRelevant
```

- Taula de les aerolínies amb més endarreriments acumulats
